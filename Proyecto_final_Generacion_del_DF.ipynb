````python
def generar_idea_ia(tipo_negocio, tipo_contenido, publicos_objetivos, tipo_objetivo):
    if cliente is None:
        return "La clave de OpenAI no está configurada. No se puede generar la idea."

    prompt = f"""
    Actúa como un experto en marketing digital. Tu tarea es generar un texto publicitario breve (máx. 5 oraciones), creativo y original.
    El texto va a usarse para el guion de un video, para un post en redes sociales o de pie de una foto.

      Contexto:
        Tipo de negocio: {tipo_negocio}
        Enfoque del contenido: {tipo_contenido}
        Públicos objetivos: {publicos_objetivos}
        Objetivo del contenido: {tipo_objetivo}
      Requisitos:
        La idea debe ser atractiva, clara y alineada con el tipo del negocio.
        Debe conectar con el público objetivo o públicos objetivos y cumplir con el enfoque del contenido.
        Usa un tono auténtico, cercano y adaptado al tipo de negocio.
        Incluye un llamado a la acción si es relevante.

      Indicaciones:
        Evita frases genéricas, clichés o respuestas poco contextualizadas, pleonasmos y redundancias.
        Devuelve solo el texto de la publicación, sin encabezados ni explicaciones.
    """

    try:
        start_time = time.time()
        response = cliente.chat.completions.create(
            model="gpt-4-turbo",
            messages=[
                {"role": "user", "content": prompt}
            ],
            max_tokens=160,  # Limitar la longitud de la idea
            n=1,            # Generar una sola idea
            stop=None,      # No especificar un stop sequence particular
            temperature=0.7, # Un valor entre 0 y 1. Mayor valor -> más aleatorio
        )
        end_time = time.time()
        generation_time = end_time - start_time
        idea = response.choices[0].message.content.strip()
        return idea, generation_time
    except Exception as e:
        return f"Error al generar la idea: {e}", None
````

````python
def limpiar_bloque_json(texto):
    # Busca el bloque entre ```json y ```
    patron = r"```json\s*(\{.*?\})\s*```"
    match = re.search(patron, texto, re.DOTALL)
    if match:
        return match.group(1)
    else:
        # Si no encuentra, intenta sacar solo la parte que parece JSON
        json_start = texto.find("{")
        json_end = texto.rfind("}") + 1
        if json_start != -1 and json_end != -1:
            return texto[json_start:json_end]
        else:
            return texto  # fallback, devuelve todo
````

````python
def evaluar_idea_ia_gpt(fila):
    prompt_eval = f"""
Eres un experto en marketing digital y redacción publicitaria. Tu tarea es evaluar una idea generada por una IA para una publicación en redes sociales. Evalúa cada criterio de forma objetiva en una escala del 0 al 5, donde:

- 0 = completamente ausente
- 1 = muy deficiente
- 2 = deficiente
- 3 = aceptable
- 4 = bueno
- 5 = excelente

### Parámetros de entrada:
- Tipo de negocio: {fila['tipo_negocio']}
- Tipo de contenido: {fila['tipo_contenido']}
- Público objetivo: {fila['publico_objetivo']}
- Objetivo del contenido: {fila['objetivo_contenido']}

### Texto generado por IA:
"{fila['idea']}"

### Criterios de evaluación:
1. **Claridad**: ¿El mensaje se comunica de forma clara, sin ambigüedades ni confusión?
2. **Alineación con el negocio**: ¿El contenido refleja adecuadamente la naturaleza del negocio?¿Está contextualizada al tipo de producto o servicio?
3. **Conexión con el público objetivo**: ¿La idea utiliza lenguaje, referencias o tono adecuados al público definido?¿Apela a intereses, emociones o necesidades del público?
4. **Cumplimiento del enfoque del contenido**: ¿La idea es coherente con el tipo de contenido indicado (ej. promocional, educativo)?¿La idea cumple con el objetivo planteado?
5. **Tono auténtico y adaptado**: ¿Se percibe una voz natural, cercana, adecuada al tipo de negocio?¿Evita frases genéricas o forzadas?
6. **Llamado a la acción (si corresponde)**: ¿Existe un llamado relevante, específico y útil para motivar acción por parte del usuario si corresponde?
7. **Originalidad y precisión**: ¿Evita frases genéricas, clichés, pleonasmos y redundancias innecesarias?

### Formato de respuesta:
Retorna únicamente la respuesta en formato JSON que tenga la siguiente estructura:
{{
  "claridad": int,
  "alineacion_negocio": int,
  "conexion_publico": int,
  "enfoque_contenido": int,
  "tono": int,
  "llamado_accion": int,
  "originalidad_precision": int,
  "puntaje_global": float  # promedio de los 7 anteriores
}}
"""
    try:
        response = cliente.chat.completions.create(
            model="gpt-4-turbo",
            messages=[{"role": "user", "content": prompt_eval}],
            max_tokens=300,
            temperature=0.2,
        )
        contenido = response.choices[0].message.content.strip()
        # Limpieza robusta
        contenido_limpio = limpiar_bloque_json(contenido)
        #print("Respuesta de GPT:\n", contenido_limpio)
        evaluacion = json.loads(contenido_limpio)
        return evaluacion

    except Exception as e:
        return {"error": str(e)}
````

````python
try:
    gemini_api_key = userdata.get('GEMINI_KEY_PROFE')
    genai.configure(api_key=gemini_api_key)
    print("Clave de Gemini cargada exitosamente.")
except Exception as e:
    print(f"Error al cargar la clave de Gemini: {e}")
    print("Asegúrate de haber guardado tu clave de Gemini en Colab (Panel izquierdo -> Secreto).")
    genai = None

#'gemini-1.5-flash-latest' o 'gemini-1.0-pro'
if genai:
    modelo_gemini = genai.GenerativeModel('gemini-1.5-flash-latest') # O el modelo que prefieras
else:
    modelo_gemini = None

def evaluar_idea_ia_gemini(fila):
    if not modelo_gemini:
        return {"error": "Modelo de Gemini no inicializado."}

    prompt_eval = f"""
Eres un experto en marketing digital y redacción publicitaria. Tu tarea es evaluar una idea generada por una IA para una publicación en redes sociales. Evalúa cada criterio de forma objetiva en una escala del 0 al 5, donde:

- 0 = completamente ausente
- 1 = muy deficiente
- 2 = deficiente
- 3 = aceptable
- 4 = bueno
- 5 = excelente

### Parámetros de entrada:
- Tipo de negocio: {fila['tipo_negocio']}
- Tipo de contenido: {fila['tipo_contenido']}
- Público objetivo: {fila['publico_objetivo']}
- Objetivo del contenido: {fila['objetivo_contenido']}

### Texto generado por IA:
"{fila['idea']}"

### Criterios de evaluación:
1. **Claridad**: ¿El mensaje se comunica de forma clara, sin ambigüedades ni confusión?
2. **Alineación con el negocio**: ¿El contenido refleja adecuadamente la naturaleza del negocio?¿Está contextualizada al tipo de producto o servicio?
3. **Conexión con el público objetivo**: ¿La idea utiliza lenguaje, referencias o tono adecuados al público definido?¿Apela a intereses, emociones o necesidades del público?
4. **Cumplimiento del enfoque del contenido**: ¿La idea es coherente con el tipo de contenido indicado (ej. promocional, educativo)?¿La idea cumple con el objetivo planteado?
5. **Tono auténtico y adaptado**: ¿Se percibe una voz natural, cercana, adecuada al tipo de negocio?¿Evita frases genéricas o forzadas?
6. **Llamado a la acción (si corresponde)**: ¿Existe un llamado relevante, específico y útil para motivar acción por parte del usuario si corresponde?
7. **Originalidad y precisión**: ¿Evita frases genéricas, clichés, pleonasmos y redundancias innecesarias?

### Formato de respuesta:
Retorna únicamente la respuesta en formato JSON que tenga la siguiente estructura:
{{
  "claridad": int,
  "alineacion_negocio": int,
  "conexion_publico": int,
  "enfoque_contenido": int,
  "tono": int,
  "llamado_accion": int,
  "originalidad_precision": int,
  "puntaje_global": float  # promedio de los 7 anteriores
}}
"""
    try:
        response = modelo_gemini.generate_content(prompt_eval)
        contenido = response.text.strip()
        # Limpieza robusta
        contenido_limpio = limpiar_bloque_json(contenido)
        #print("Respuesta de Gemini:\n", contenido_limpio)
        evaluacion = json.loads(contenido_limpio)
        return evaluacion

    except Exception as e:
        return {"error": str(e)}
````

````python
def validar_json_eval(json_eval):
    criterios = {
        "claridad": int,
        "alineacion_negocio": int,
        "conexion_publico": int,
        "enfoque_contenido": int,
        "tono": int,
        "llamado_accion": int,
        "originalidad_precision": int,
        "puntaje_global": float
    }
    if not isinstance(json_eval, dict):
        return False
    # Convert the keys of both dictionaries to sets before using issubset
    return set(criterios.keys()).issubset(set(json_eval.keys()))
````

````python
def sumar_puntajes_globales(df):
    total = 0
    contador = 0

    for index, row in df.iterrows():
        for col in ["evaluacion_gpt", "evaluacion_gemini"]:
            evaluacion = row[col]
            if isinstance(evaluacion, dict) and "puntaje_global" in evaluacion:
                try:
                    total += float(evaluacion["puntaje_global"])
                    contador += 1
                except:
                    continue  # si hay error al convertir el número, lo ignora

    return total, contador, total / contador if contador > 0 else 0
````

````python
def aprobar_respuesta(df):
    for index, row in df.iterrows():
        puntajes = []
        for col in ["evaluacion_gpt", "evaluacion_gemini"]:
            evaluacion = row[col]
            if isinstance(evaluacion, dict) and "puntaje_global" in evaluacion:
                try:
                    puntaje = float(evaluacion["puntaje_global"])
                    puntajes.append(puntaje)
                except:
                    continue  # ignora errores de conversión

        if puntajes:
            promedio = sum(puntajes) / len(puntajes)
            df.at[index, "aprobado"] = promedio >= 4.1
        else:
            df.at[index, "aprobado"] = False  # si no hay ningún puntaje válido

    return df
````

````python
df_entrenamiento = aprobar_respuesta(dataframe_entrenamiento)
df_entrenamiento = df_entrenamiento.drop('evaluacion_valida', axis=1)
df_entrenamiento
````
